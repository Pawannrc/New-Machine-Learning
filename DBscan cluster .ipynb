```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
from sklearn.datasets import make_moons             # make_moom This dataset inbuilt in scikit learn we import for practical 
```

```python
x,y=make_moons(n_samples=250, noise=0.05)
```

```python
df={'data 1':x[:,0],'data2':x[:,1]}
```

```python
data=pd.DataFrame(df)
```

```python
data.head(2)
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x='data 1',y='data2',data=data)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
db=DBSCAN(eps=0.2,min_samples=5)
data['predict']=db.fit_predict(data)              # eps otherword called epsillon mean radius of circle 
                                                # min_samples meaning how you want circle
```

```python
data.head(2)                                   # this is predicted dataset
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x='data 1',y='data2',data=data,hue='predict')
plt.show()
```
