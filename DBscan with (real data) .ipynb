```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('moon.csv')
```

```python
data.head(3)
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x='data_x',y='data_x2',data=data)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
db=DBSCAN( eps=0.2, min_samples=5)
data['predict']=db.fit_predict(data)
```

```python
data.head(2)
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x='data_x',y='data_x2',data=data,hue='predict')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
ss=[]
no_c=[r for r in range(2,21):]
db1= DBSCAN(  eps=0.2,min_samples=5)
db1.fit(data)
ss.append(silhouette_score(data,db1.labels_))
```

```python
plt.plot(ss,data=data)
plt.show()
```
