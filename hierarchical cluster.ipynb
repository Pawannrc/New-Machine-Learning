```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('raw iris.csv')
data.head(3)
```

```python
#sns.pairplot(data=data)
#plt.show                             checking here through pairplot what kind of data is linear or non linear 
```

```python
import scipy.cluster.hierarchy as sc
```

```python
plt.figure(figsize=(15,8))
sc.dendrogram(sc.linkage(data,method='single', metric='euclidean'))       # dendrogram graph through we find no of group means n_cluster
plt.show()                                                              # method is type of hyper parameter
                                                                        # Here we found 2 group through dendrograph
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac=AgglomerativeClustering(n_clusters=2,linkage='single')      # n_cluster mean number of group aslo called cluster
data['predict']=ac.fit_predict(data)
```

```python
data.head(2)
```

```python
sns.pairplot(data=data,hue='predict')
plt.show()
```
