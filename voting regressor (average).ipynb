```python
import pandas as pd
```

```python
data=pd.read_csv('salary_dataset.csv')
```

```python
data.head(2)
```

```python
data.drop(columns="Unnamed: 0",inplace=True)
```

```python
data.head(2)
```

```python
x=data[['YearsExperience']]
y=data['Salary']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
```

```python
dc=DecisionTreeRegressor()
dc.fit(x_train,y_train)
dc.score(x_train,y_train)*100, dc.score(x_test,y_test)*100
```

```python
sv=SVR()
sv.fit(x_train,y_train)
sv.score(x_train,y_train)*100, sv.score(x_test,y_test)*100
```

```python
knn=KNeighborsRegressor()
knn.fit(x_train,y_train)
knn.score(x_train,y_train)*100, knn.score(x_test,y_test)*100
```

# vs

```python
from sklearn.ensemble import VotingRegressor
```

```python

model=[ ('dc1',DecisionTreeRegressor()),
        ('sv1',SVR()),                                               
        ('knn1',KNeighborsRegressor()) ]                                 
    
vr=VotingRegressor(model,weights=[13,26,52])
vr.fit(x_train,y_train)
vr.score(x_train,y_train)*100,vr.score(x_test,y_test)*100
```

# here it is  models vs votingregressor prediction comparison (Average)

```python
prd={'dc':dc.predict(x_test),'sv':sv.predict(x_test),'knn':knn.predict (x_test),'vr':vr.predict(x_test)}
```

```python
data=pd.DataFrame(prd)
```
