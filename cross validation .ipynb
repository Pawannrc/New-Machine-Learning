```python
import pandas as pd
```

```python
data=pd.read_csv('moondataset.csv')
```

```python
data.head(3)
```

```python
x=data.iloc[:,:-1]
```

```python
y=data['label']
```

```python
from sklearn.linear_model import LogisticRegression # we use logistic because before cross validation you say what algo you use
```

```python
from sklearn.model_selection import cross_val_score ,KFold ,StratifiedKFold 
```

```python
p1=cross_val_score(LogisticRegression(),x,y,cv=5) # cv mean how many times you train model you can put according to you
```

```python
p1.sort()
p1*100
```

```python
p2=cross_val_score(LogisticRegression(),x,y,cv=StratifiedKFold(n_splits=10)) # StratifiedKFold only work in unbalance data like logistic
```

```python
p2.sort()
p2*100
```

```python
p3=cross_val_score(LogisticRegression(),x,y,cv=KFold(n_splits=10)) # kfold use both linear or logistic
```

```python
p3.sort()
p3*100
```
